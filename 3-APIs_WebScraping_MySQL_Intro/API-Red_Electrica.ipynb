1. Extracción de Datos a Nivel Nacional:
- Extraer datos de generación de energía renovable y no renovable a nivel nacional desde el **1 de enero de 2011 hasta el 31 de diciembre de 2022**. La información debe ser recopilada a nivel **mensual**.
- Transformar los datos para incluir una columna nueva que identifique si la energía es renovable o no renovable.

https://apidatos.ree.es/es/datos/generacion/estructura-generacion?start_date=2014-01-01T00:00&end_date=2018-12-31T23:59&time_trunc=year&geo_trunc=electric_system&geo_limit=ccaa&geo_ids=7

```python
#1. 
json_generation = {}

for year in range(12):
    year += 2011
    print(year)
    
    url = f"https://apidatos.ree.es/es/datos/generacion/estructura-generacion?start_date={year}-01-01T00:00&end_date={year}-12-31T23:59&time_trunc=month"
    print(url)
    llamada = requests.get(url)
    print(llamada)

    llamada.status_code
    json_generation[year] = llamada.json()
    
```

```python
print(len(json_generation[year]["included"]))
print(json_generation[year]["included"]) #Es una lista de 18 elementos
```

```python
json_generation[year]["included"][0].keys()
```

```python
diccionario_renovables_no_renovables = {"tipo_energia": [],
                             "tipo_generacion":[],
                             "fecha": [],
                             "valor": []}
meses = 0

diccionario_renovables_no_renovables = []

for year in range(2011, 2023):
    for energia in json_generation[year]["included"]:
        for mes in range(len(energia["attributes"]["values"])):
            diccionario_temporal = {}
            diccionario_temporal["tipo_energia"] = energia["type"]
            diccionario_temporal["tipo_generacion"] = energia["attributes"]["type"]
            diccionario_temporal["fecha"] = energia["attributes"]["values"][mes]["datetime"]
            diccionario_temporal["valor"] = energia["attributes"]["values"][mes]["value"]

            diccionario_renovables_no_renovables.append(diccionario_temporal)
```

```python
#2.
#Inicializamos variables y definimos diccionarios

cod_comunidades = {'Ceuta': 8744,
                'Melilla': 8745,
                'Andalucía': 4,
                'Aragón': 5,
                'Cantabria': 6,
                'Castilla - La Mancha': 7,
                'Castilla y León': 8,
                'Cataluña': 9,
                'País Vasco': 10,
                'Principado de Asturias': 11,
                'Comunidad de Madrid': 13,
                'Comunidad Foral de Navarra': 14,
                'Comunitat Valenciana': 15,
                'Extremadura': 16,
                'Galicia': 17,
                'Illes Balears': 8743,
                'Canarias': 8742,
                'Región de Murcia': 21,
                'La Rioja': 20}
                
diccionario_ccaa = {"tipo_energia": [],
                    "tipo_generacion":[],
                    "fecha": [],
                    "valor": [],
                    "ccaa": []}

mes = 0

diccionario_ccaa = []
```

```python
print(cod_comunidades.values())
```

```python
#2. 
for comunidad in cod_comunidades.values():
    print(comunidad)
    for years in range(2012,2023):
        
        url2 = f"https://apidatos.ree.es/es/datos/generacion/estructura-generacion?start_date={years}-01-01T00:00&end_date={years}-12-31T23:59&time_trunc=month&geo_limit=ccaa&geo_ids={comunidad}"
        
        result = requests.get(url2)
        
        print(result.status_code)

        result_json = result.json()
        for energy in result_json["included"]:
            for mes in range(len(energy["attributes"]["values"])):
                temporal_dicc = {}
                temporal_dicc["tipo_energia"] = energy["type"]
                temporal_dicc["tipo_generacion"] = energy["attributes"]["type"]
                temporal_dicc["fecha"] = energy["attributes"]["values"][mes]["datetime"]
                temporal_dicc["valor"] = energy["attributes"]["values"][mes]["value"]
                ccaa = [k for k, v in cod_comunidades.items() if v == comunidad]
                temporal_dicc["ccaa"] = ccaa

                diccionario_ccaa.append(temporal_dicc)
```

```python
tabla_ccaa = pd.DataFrame(diccionario_ccaa)	
tabla_ccaa
```
